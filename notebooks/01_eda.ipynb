# NodeSense Exploratory Data Analysis

Initial look at the CICIDS-2018 dataset: class balance, feature distributions, and data quality issues to handle in preprocessing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../data/cicids2018.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Class balance
df['Label'].value_counts()
```

```python
# Data quality: check for infinities and nulls from CICFlowMeter
df.replace([np.inf, -np.inf], np.nan, inplace=True)
print('Null counts per column:')
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Feature distributions for a few key columns
cols = ['Flow Duration', 'Flow Bytes/s', 'Fwd Packet Length Mean']
df[cols].hist(bins=50, figsize=(12, 4))
plt.tight_layout()
plt.show()
```
